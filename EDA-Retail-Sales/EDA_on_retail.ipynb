# Exploratory Data Analysis (EDA) on Retail Sales Dataset

## Introduction
Exploratory Data Analysis (EDA) is the process of analyzing datasets to summarize their main characteristics using statistical methods and visualizations.

In this project, we analyze a retail sales dataset containing information about customers, product categories, and transaction details. The objective is to discover patterns, trends, and insights that can help businesses understand customer behavior and sales performance.

## Importing Required Libraries
The following libraries are used for data analysis and visualization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading the Dataset
The dataset is loaded using pandas, and the first few rows are displayed to understand the structure.

```python
df = pd.read_csv("retail_sales_dataset.csv")
df.head()
```

## Dataset Overview
We check the size of the dataset and the data types of each column.

```python
df.shape
```

```python
df.info()
```

## Checking Missing Values
Missing values can affect analysis results, so we check whether any columns contain null values.

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

## Descriptive Statistics
Descriptive statistics summarize the main features of numerical variables such as age, quantity, price, and total sales.

```python
df.describe()
```

## Age Distribution of Customers
This histogram shows the distribution of customer ages in the dataset.

```python
plt.figure()
df["Age"].hist(edgecolor='black')
plt.title("Age Distribution of Customers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

## Observation:
The histogram shows how customers are distributed across different age groups.

## Sales by Product Category
We analyze total sales generated by each product category.

```python
category_sales = df.groupby("Product Category")["Total Amount"].sum()
category_sales
```

```python
category_sales.plot(kind="bar")
plt.title("Total Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.show()
```

## Observation:
The bar chart shows which product category generates the highest total sales.

## Gender Distribution of Customers
This chart shows how customers are distributed by gender.

```python
df["Gender"].value_counts().plot(kind="bar")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

## Monthly Sales Trend
We analyze how sales change across different months.

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Month"] = df["Date"].dt.month
```

```python
monthly_sales = df.groupby("Month")["Total Amount"].sum()
monthly_sales
```

```python
monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

## Observation:
The line chart shows how sales fluctuate across different months of the year.

## Correlation Analysis
A heatmap is used to examine relationships between numerical variables.

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

## Key Insights

1. Electronic category generated higher sales compared to others.
2. Customers aged 60+ make the most purchases..
3. Sales are higher in the month of June.
4. Female customers purchased slightly more products than male customer.
