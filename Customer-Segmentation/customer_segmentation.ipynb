# Customer Segmentation Analysis

This project performs customer segmentation using clustering techniques. 
The goal is to group customers with similar purchasing behavior to help businesses 
target marketing strategies effectively.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("ifood_df.csv")
df.head()
```

## Data Cleaning

Checking for missing values and removing unnecessary columns.

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

## Feature Selection

For segmentation we select variables related to customer spending.

```python
features = df[['Income','MntWines','MntFruits','MntMeatProducts',
                'MntFishProducts','MntSweetProducts','MntGoldProds']]

features.head()
```

## Feature Scaling

Scaling ensures all features contribute equally to clustering.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_data = scaler.fit_transform(features)
```

## Determining Optimal Number of Clusters

The Elbow Method is used to identify the optimal number of clusters for K-Means clustering.

```python
from sklearn.cluster import KMeans

wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)

plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")

plt.show()
```

## Applying K-Means Clustering

Using the optimal number of clusters obtained from the Elbow Method.

```python
kmeans = KMeans(n_clusters=4, random_state=42)

df['Cluster'] = kmeans.fit_predict(scaled_data)
```

## Customer Segmentation Visualization

```python
plt.scatter(df['Income'], df['MntWines'], c=df['Cluster'])

plt.xlabel("Income")
plt.ylabel("Wine Spending")
plt.title("Customer Segments")

plt.show()
```

Analyzing the characteristics of each customer segment.
